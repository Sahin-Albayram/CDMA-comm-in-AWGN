```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv("scenario_1.csv")
df1.drop('Unnamed: 0',axis='columns',inplace=True)
df2 = pd.read_csv("scenario_2.csv")
df2.drop('Unnamed: 0',axis='columns',inplace=True)
```

```python
plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1)
plt.title("Semilogx")
for i in range(3):
    data1 = df1.iloc[i,:]
    data2 = df2.iloc[i,:]
    plt.semilogx(data1,label = f"node{i} scenario_1")
    plt.semilogx(data2,label = f"node{i} scenario_2")
plt.xlabel("Noise std values")
plt.ylabel("BRE")
plt.legend()
plt.subplot(1, 3, 2)
plt.title("Semilogy")
for i in range(3):
    data1 = df1.iloc[i,:]
    data2 = df2.iloc[i,:]
    plt.semilogy(data1,label = f"node{i} scenario_1 ")
    plt.semilogy(data2,label = f"node{i} scenario_2")
plt.xlabel("Noise std values")
plt.ylabel("BRE")
plt.legend()
plt.subplot(1, 3, 3)
plt.title("loglog")

for i in range(3):
    data1 = df1.iloc[i,:]
    data2 = df2.iloc[i,:]
    plt.loglog(data1,label = f"node{i} scenario_1")
    plt.loglog(data2,label = f"node{i} scenario_2")
plt.xlabel("Noise std values")
plt.ylabel("BRE")
plt.legend()
plt.show()
```
